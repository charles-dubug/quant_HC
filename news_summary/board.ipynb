```python
# 读取JSON文件
with open('f:/AI/news-summary/industry_board_data.json', 'r', encoding='utf-8') as f:
    industry_data = json.load(f)

# 提取所有板块名称
industry_names = [item['板块名称'] for item in industry_data]

# 打印结果
print(industry_names)

# 如果需要保存到文件
with open('f:/AI/news-summary/industry_boards.py', 'w', encoding='utf-8') as f:
    f.write("industry_names = " + str(industry_names))
```
